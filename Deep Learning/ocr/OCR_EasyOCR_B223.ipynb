```python
import easyocr
import cv2
```

```python
reader = easyocr.Reader(['tr', 'en'], gpu=False)  #Hindi, telugu, and English
```

**reader = easyocr.Reader(['tr', 'en'], gpu=False)**: easyocr ktphanesinden bir okuyucu oluturuyoruz. Bu okuyucu, Trke ve ngilizce dillerini tanyacak ekilde ayarlanmtr (['tr', 'en']). 

**gpu=False** parametresi, grafik ilemciyi kullanmayacamz belirtir, bu nedenle ilemci tabanl bir zm kullanlr.

```python
img = cv2.imread('gazi.jpg')
```

```python
results = reader.readtext(img, detail=1, paragraph=True) 
```

**results_paragraph = reader.readtext(img, detail=1, paragraph=True):** Oluturduumuz okuyucu kullanarak resimdeki metni okur. 

**detail=1** detayl sonular almay salar ve 

**paragraph=True** ile metin paragraf olarak alnr.
