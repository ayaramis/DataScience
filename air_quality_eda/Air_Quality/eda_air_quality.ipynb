4th Max: This refers to the fourth highest value in a dataset. It can be used to identify extreme events or outliers without focusing solely on the absolute maximum, which might be atypical.

Weighted Annual Mean: This is a calculation where the mean of a dataset is weighted based on certain factors, typically time periods. For example, the annual mean of a pollutant might be weighted by the number of days in each month to give a more accurate representation of exposure over the entire year.

98th Percentile: This is the value below which 98% of the data points fall. It gives an idea of the upper end of the typical range of values, excluding extreme outliers. Similarly, the 99th Percentile would be the value below which 99% of the data points fall.

2nd Max: This simply refers to the second highest value in a dataset. It can be used along with the maximum to understand the distribution of extreme values.

Annual Mean: This is the average of all data points within a year. It provides a simple and overall picture of the typical value throughout the year.

Max 3-Month Average: This is the highest average value calculated over any three consecutive months within a year. It helps identify periods of relatively high exposure compared to the whole year.

These metrics are used in various ways depending on the specific context. For example, air quality standards might be based on annual means or percentiles, while risk assessments might consider extreme events like the 4th max or the max 3-month average.
